```python
import numpy as np
import random

def create_maze(grid_size):    
    grid = np.ones((grid_size, grid_size), dtype=int)

    start_x, start_y = (1, 1)
    # grid[start_x, start_y] = 2

    directions = [(0, 2), (2, 0), (0, -2), (-2, 0)]
    
    def carve(x, y):
        
        random.shuffle(directions)
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 < nx < grid_size and 0 < ny < grid_size and grid[nx, ny] == 1:

                grid[nx, ny] = 0
                grid[x + dx // 2, y + dy // 2] = 0
                carve(nx, ny)

    carve(start_x, start_y)
    
    grid[:, 0] = 0
    #grid[0, :] = 0
   
    return grid

grid_size = 100
    
maze_grid = create_maze(grid_size)

for row in maze_grid:
    print(' '.join(['*' if cell == 1 else '$' if cell == 2 else ' ' for cell in row]))
```

```python
OBSTACLES = []
for row in range(len(maze_grid)):
    for col in range(len(maze_grid)):
        if maze_grid[row][col] == 1:
            OBSTACLES.append((row,col))
OBSTACLES
```

```python
GRID_SIZE = 100
START = (0,0)
GOAL = (GRID_SIZE-1,GRID_SIZE-1)
#OBSTACLES = [(0,0),(0,2),(0,4),(1,4),(2,1),(2,2),(3,4),(4,1),(4,2),(4,4)]
ACTIONS = [(-1,0),(0,1),(1,0),(0,-1)]
EPSILON = 0.3
ALPHA = 0.3
GAMMA = 0.99
EPISODES = 10000
QTable = np.zeros((GRID_SIZE,GRID_SIZE,len(ACTIONS)))
```

```python
def choose_action(state, QTable):
    prob = random.uniform(0,1)
    if prob < EPSILON:
        return random.choice(ACTIONS)
    else:
        return ACTIONS[np.argmax(QTable[state[0]][state[1]])]
```

```python
def is_valid_state(state):
    if (0 <= state[0] < GRID_SIZE and 
        0 <= state[1] < GRID_SIZE and 
        state not in OBSTACLES):
        return True
```

```python
def get_next_state(state, action):
    next_state = (state[0]+action[0], state[1]+action[1])
    if is_valid_state(next_state):
        return next_state 
    else:
        return state
```

```python
def get_reward(state, next_state):
    if next_state == GOAL:
        return 100
    elif next_state in OBSTACLES or next_state == state:
        return -10
    else:
        return -1
```

```python
def update_q_table(QTable, state, action, reward, next_state):
    action_idx = ACTIONS.index(action)
    QTable[state[0]][state[1]][action_idx] += ALPHA * (reward + GAMMA * np.max(QTable[next_state[0]][next_state[1]]) - QTable[state[0]][state[1]][action_idx])
```

```python
def agent_is_learning():
    for _ in range(EPISODES):
        state = START
        while state != GOAL:
            action = choose_action(state, QTable)
            next_state = get_next_state(state, action)
            reward = get_reward(state, next_state)
            update_q_table(QTable, state, action, reward, next_state)
            state = next_state
        print("Epoch ",_+1," done.")
    return QTable
```

```python
QTable = agent_is_learning()
```

```python
def visualize_q_table_as_grid(q_table: np.ndarray) -> None:
    """Visualize the Q-table as a grid with all action values for each state."""
    action_symbols = ['^', '>', 'v', '<']
    
    print("\nDetailed Q-table Grid:")
    
    # Header
    header = "   |" + "|".join(f"   ({i},{j})   " for i in range(GRID_SIZE) for j in range(GRID_SIZE)) + "|"
    print(header)
    print("-" * len(header))

    for action_idx, action_symbol in enumerate(action_symbols):
        row = f" {action_symbol} |"
        for i in range(GRID_SIZE):
            for j in range(GRID_SIZE):
                if (i, j) == GOAL:
                    cell = "   GOAL    "
                elif (i, j) in OBSTACLES:
                    cell = " OBSTACLE  "
                else:
                    q_value = q_table[i, j, action_idx]
                    cell = f" {q_value:9.2f} "
                row += cell + "|"
        print(row)
        print("-" * len(header))

def visualize_best_actions_grid(q_table: np.ndarray) -> None:
    """Visualize the best action and its Q-value for each state in a grid."""
    action_symbols = ['^', '>', 'v', '<']
    
    print("\nBest Actions Grid:")
    header = "-" * (14 * GRID_SIZE + 1)
    print(header)

    for i in range(GRID_SIZE):
        row = "| "
        for j in range(GRID_SIZE):
            if (i, j) == GOAL:
                cell = "   GOAL    "
            elif (i, j) in OBSTACLES:
                cell = " OBSTACLE  "
            else:
                best_action_idx = np.argmax(q_table[i, j])
                best_q_value = q_table[i, j, best_action_idx]
                cell = f"{action_symbols[best_action_idx]}:{best_q_value:7.2f}  "
            row += cell + " | "
        print(row)
        print(header)
```

```python
# Visualize the Q-table as a grid
visualize_q_table_as_grid(QTable)
```

```python
# Visualize the best actions and their Q-values in a grid
visualize_best_actions_grid(QTable)
```
